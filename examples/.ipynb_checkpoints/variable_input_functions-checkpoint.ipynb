````julia
using PETLION, Plots
````

````julia
p = petlion(LCO;
    jacobian=:symbolic
    );
````

# Variable current/power/voltage guide

All operating modes to PETLION (`I`, `V`, and `P`) accept constants or functions as inputs. Say you have some custom current protocol `I_fun` that you want to run with
```julia
sol = simulate(p, 1000, I=I_fun)
```
You can set it up four different ways:
1. A function of time
```julia
I_fun(t) = ...
```
2. A function of time and the parameters
```julia
I_fun(t,p) = ...
```
3. A function of time, parameters, and algebraic states
```julia
I_fun(t,Y,p) = ...
```
4. A function of time, parameters, algebraic states, and differential states
```julia
I_fun(t,Y,YP,p) = ...
```

The last two can be helpful for model predictive control purposes.



**Important Info**
<ul>
<li>All times start at t = 0, even if the simulation follows another one</li>
<li>Discontinuities are allowed (see examples)</li>
<li>The functions may be discontinuous in time, but performance is best with continuous functions</li>
</ul>

# Example 1: Discontinuous current function

Let's run a multi-constant current segment

````julia
function I_fun1(t)
    if t < 100
        return 1
    else
        return 0.5
    end
end
````

Right now, the model is unaware of the discontinuity at t = 100. Here's the compiled time for the simulation.

````julia
@time sol = simulate(p,200,I=I_fun1,SOC=0)
````

We can specify any number of discontinuities using the keyword argument `tdiscon`, which accepts a vector of numbers as an input. After specifying when the discontinuity occurs

````julia
@time sol = simulate(p,200,I=I_fun1,SOC=0,tdiscon=[100])
````

Much better!

# Example 2: Current Ramp

We can add new parameters to our functions as well so we don't have to deal with compilation time

````julia
I_ramp(t,p) = p.Î¸[:ramp_val]*t
````

Let's try a ramp of $$ I(t) = \frac{t}{100} .$$

````julia
p.Î¸[:ramp_val] = 1/100
@time sol_ramp1 = simulate(p,100,I=I_ramp,SOC=0)
println(sol_ramp1)
plot(sol_ramp1)
````

Now, let's try a ramp of $$ I(t) = \frac{t}{10} .$$

````julia
p.Î¸[:ramp_val] = 1/10
@time sol_ramp2 = simulate(p,100,I=I_ramp,SOC=0)
println(sol_ramp2)
plot(sol_ramp2)
````

No extra compilation cost!

# Example 3: Variable power and voltage

````julia
P_fun(t) = sin(t)
V_fun(t) = 3.5 + 0.1cos(t)
````

````julia
@time sol_P_fun = simulate(p,10,P=P_fun,SOC=0.5,outputs=:all)
plot(sol_P_fun, :P)
````

````julia
@time sol_V_fun = simulate(p,10,V=V_fun,SOC=0.5,outputs=:all)
plot(sol_V_fun, :V)
````
